```python
lando_laps = suz_laps[(suz_laps['Driver'] == "NOR") & (suz_laps['PitOutTime'].isna()) & (suz_laps['PitInTime'].isna())
                      & (suz_laps['TrackStatus'] == '1')]
display(lando_laps)
```

```python
plt.plot(lando_laps['LapNumber'], lando_laps['LapTime'].dt.total_seconds())
plt.show()
```

```python
race_lap_count = {}

for index, event in schedule_24.iterrows():
    race = event['Location']
    session = fastf1.get_session(2024, race, 'R')
    session.load()
    laps = session.laps
    race_lap_count[race] = laps['LapNumber'].max()
```

```python
fuel_consumption = pd.DataFrame(columns=['Track', 'LapCount', 'FuelUsePerLap', 'LapTimeGained', 'TotalTimeGained'])

for key, value in race_lap_count.items():
    fuel_burn = round(110 / value, 2)
    time_gained = round((110 / value) * 0.03, 2)
    total_gained = time_gained * value
    fuel_consumption.loc[len(fuel_consumption)] = [key, value, fuel_burn, time_gained, total_gained]
    
display(fuel_consumption)
```

```python
lando_df = pd.DataFrame(lando_laps)
corrected_lando = pd.DataFrame(columns=['LapNumber', 'LapTime', 'CorLapTime', 'WeightFactor', 'Stint'])
filtered = lando_df[['LapNumber', 'LapTime', 'Stint']]

for key, value in filtered.iterrows():
    weight_factor = value['LapNumber'] * 0.06
    cor_lap = weight_factor + value['LapTime'].total_seconds()
    corrected_lando.loc[len(corrected_lando)] = [value['LapNumber'], value['LapTime'].total_seconds(), cor_lap, weight_factor, value['Stint']]

plt.plot(corrected_lando['LapNumber'], corrected_lando['CorLapTime'])
plt.show()
```

```python
display(corrected_lando)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Filter for a specific stint (if you have multiple stints)
stint_df = corrected_lando[corrected_lando['Stint'] == 1].copy()  # Assuming you have multiple stints

X = stint_df[['LapNumber']]
y = stint_df['CorLapTime']

# Fit linear model
model = LinearRegression()
model.fit(X, y)

base_time = model.intercept_
degradation_rate = model.coef_[0]

print(f"Base Lap Time: {base_time:.3f} s")
print(f"Tire Degradation Rate: {degradation_rate:.4f} s per lap")

# Plot the data
plt.figure(figsize=(10,6))
plt.scatter(X, y, label='Corrected Lap Time')
plt.plot(X, model.predict(X), color='red', label='Linear Fit')
plt.xlabel('Lap Number')
plt.ylabel('Corrected Lap Time (s)')
plt.title('Tire Degradation Model')
plt.legend()
plt.grid(True)
plt.show()
```
